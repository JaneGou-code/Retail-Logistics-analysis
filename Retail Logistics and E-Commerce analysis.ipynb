# 2. Data Cleaning

```python
df.tail(50)
```

```python
# Find out the missing values
df =df .replace(r'^\s*$',np.nan, regex=True)
df.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

def fill_missing_values_with_rf(df, to_fill):
    df_copy = df.copy()
    
    # separate training data and testing data
    train_data = df_copy.dropna(subset=[to_fill])
    missing_data = df_copy[df_copy[to_fill].isnull()]
    

    y_train = train_data[to_fill]
    X_train = train_data.drop(columns=[to_fill])
    X_missing = missing_data.drop(columns=[to_fill])
    

    num_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
    cat_features = X_train.select_dtypes(include=['object']).columns.tolist()
    
    # Impute mean value as the filling method
    num_transformer = SimpleImputer(strategy='mean')
    cat_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])
    
    # ColumnTransformer
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', num_transformer, num_features),
            ('cat', cat_transformer, cat_features)
        ])
    
    # Buiding model
    model = Pipeline(steps=[('preprocessor', preprocessor),
                            ('classifier', RandomForestRegressor(n_estimators=100, random_state=0))])
    
    model.fit(X_train, y_train)
    
    # Predict the missing data
    df_copy.loc[df_copy[to_fill].isnull(), to_fill] = model.predict(X_missing)
    
    return df_copy

# Filling
df_filled = fill_missing_values_with_rf(df, 'shoppostcode')
df_filled = fill_missing_values_with_rf(df_filled, 'lonadj')
df_filled = fill_missing_values_with_rf(df_filled, 'latadj')
```

```python
# Check the missing values again
df_filled =df_filled .replace(r'^\s*$',np.nan, regex=True)
df_filled.isnull().sum()
```

# 3. Data transfer

```python
df_grouped = df_filled.groupby(['shopgln', 'producttype']).agg({
    'qtyinshop': 'sum',
    'qtyonorder': 'sum',
    'qtysold': 'sum',
    'lonadj': 'first', # Retaining the first value for each group
    'latadj': 'first', 
    'supplierid': 'first', 
 
}).reset_index()

df_grouped
```

```python
df_dummies = pd.get_dummies(df_grouped['producttype'], prefix='producttype')
df_one = pd.concat([df_grouped, df_dummies], axis=1)
df_one = df_one.drop('producttype', axis=1) 

df_one
```

```python
df_final_grouped = df_grouped.groupby('shopgln').agg({
    'qtyinshop': 'sum',
    'qtyonorder': 'sum',
    'qtysold': 'sum',
    'lonadj': 'first',  # Retaining the first value for each group
    'latadj': 'first',  
}).reset_index()
df_final_grouped
```
